# Day 3.1

```python
with open('day3input.txt', 'r') as file:
    inputs = [line.strip() for line in file.readlines()]
transpose = [[inputs[i][j] for i in range(0,len(inputs))] for j in range(0,len(inputs[0]))]

def most_common(line):
    counts = {'1':0,'0':0}   
    for i in line: counts[i] += 1 
    return sorted(counts,key= lambda x:counts[x])[1]

gamma = ''.join(most_common(line) for line in transpose)
epsilon = ''.join('1' if i =='0' else '0' for i in list(gamma))
int(gamma,2)*int(epsilon,2)
```

# Day 3.2

```python
%%time
with open('day3input.txt', 'r') as file:
    inputs = [line.strip() for line in file.readlines()]

transpose = lambda inputs: [[inputs[i][j] for i in range(0,len(inputs))] for j in range(0,len(inputs[0]))]
columns = transpose(inputs)

def commonality(line,most=True):
    counts = {'1':0,'0':0}   
    for i in line: counts[i] += 1 
    if counts['1'] == counts['0']:
        return '1' if most else '0'
    s = sorted(counts,key= lambda x:counts[x])
    return s[1] if most else s[0]

ans = []
for most in [True, False]:
    valid = inputs.copy()
    column = 0 
    while len(valid) > 1 and column < len(inputs[0]):
        most_common = commonality(transpose(valid)[column],most)
        ok = [i for i in valid if i[column] == most_common]
        if len(ok) > 0 : valid = ok.copy()
        column +=1
    ans.append(valid[0])
int(ans[0],2)*int(ans[1],2)
```

```python
with open('day4input.txt','r') as file:
    inputs = file.readlines()
    order = [int(i) for i in inputs[0].split(',')]
    offset = 2
    boards = [] 
    while offset < len(inputs):
        boards.append([i.strip() for i in inputs[offset:offset+5]])
        offset += 6
    for i in range(0,len(boards)):
        for j in range(0,len(boards[i])):
            boards[i][j] = [int(val) for val in boards[i][j].split(' ') if val!='']

import numpy as np
boards = np.array(boards)
marked = np.zeros(shape=boards.shape)

for number in order:
    marked += boards == number  
    for ind,board in list(enumerate(marked)):
        if any(all(i) for i in board) or any(all(i) for i in board.T):
            break
    else:
        continue
    break
    
sum(boards[ind][np.where(board==0)])*number
```

# Day 4.2

```python
with open('day4input.txt','r') as file:
    inputs = file.readlines()
    order = [int(i) for i in inputs[0].split(',')]
    offset = 2
    boards = [] 
    while offset < len(inputs):
        boards.append([i.strip() for i in inputs[offset:offset+5]])
        offset += 6
    for i in range(0,len(boards)):
        for j in range(0,len(boards[i])):
            boards[i][j] = [int(val) for val in boards[i][j].split(' ') if val!='']

import numpy as np
boards = np.array(boards)
marked = np.zeros(shape=boards.shape)

def find(a,b):
    for i in range(len(a)):
        if np.all(a[i] == b):
            return i
              
for number in order:
    marked += boards == number  
    for ind,board in list(enumerate(marked)):
        if any(all(i) for i in board) or any(all(i) for i in board.T):
            if len(boards) > 1:
                boards = np.delete(boards,find(marked,board),axis=0)
                marked = np.delete(marked,find(marked,board),axis=0)
            else:
                break
    else:
        continue
    break
sum(boards[find(marked,board)][np.where(board==0)])*number
```

# Day 5.2

```python
import numpy as np

class point:
    def __init__(self,st):
        st = st.split(',')
        self.x = int(st[0].strip())
        self.y = int(st[1].strip())
        
with open('day5input.txt','r') as file:
    inputs = file.readlines()
    inputs = [i.split(' -> ') for i in inputs]
    inputs = [[point(i[0]),point(i[1].strip())] for i in inputs]

grid = np.zeros(shape=(1000,1000))

for line in inputs:
    [x1,x2] = sorted([line[1].x,line[0].x])
    [y1,y2] = sorted([line[1].y,line[0].y])
    if x1 == x2:
        grid[y1:y2+1,x1] += 1
    elif line[1].y == line[0].y:
        grid[y1,x1:x2+1] += 1     
    elif abs(line[1].x - line[0].x) == abs(line[1].y - line[0].y):
        m = (line[0].y - line[1].y) / (line[0].x - line[1].x)
        c = line[0].y - (m*line[0].x) 
        for i in range(x1,x2+1):
            y = int((m*i)+ c)
            grid[y][i] += 1         
len(grid[grid >= 2])
```

# Day 6.1

```python
%%time

with open('day6input.txt','r') as file:
    inputs = file.readlines()[0].split(',')
    inputs = [int(i) for i in inputs]


def calc(days,inputs):
    for i in range(days):
        toAdd = 0
        for ind,j in enumerate(inputs):
            if j == 0:
                toAdd += 1
            if(j < 7):
                inputs[ind] = ((j-1) % 7)
            else:
                inputs[ind] -= 1

        inputs += [8]*toAdd
    return len(inputs)
    
print(calc(80,inputs))
print(inputs.__sizeof__() /  1_000_000,'MB')
```
